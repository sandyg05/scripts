````python
# 1.A Extracting the property types from title field

# Disadvantage: I created a boolean mask to see which type is in the title string. The disadvantage of this approach is it calls contains() method for every type times every row in the data set.
# Advantage: I'm not sure this operation grows linearly because pandas dataframe does those operations in parallel, and it is pretty straightforward and readable.
def extract_type(df):
    
    property_types = {'apartment':'apartments', 'penthouse':'apartments', 'duplex':'apartments', 
                      'house':'houses', 'villa':'houses','country estate':'houses', 'moradia':'houses', 'quinta':'houses',
                      'plot':'plots', 'land':'plots'}
    
    for property_type in property_types.keys():
        type_mask = df['title'].str.contains(property_type, case=False) # Creating a boolean mask according to the type in which title contains
        # print(type_mask)
        df.loc[type_mask, 'type'] = property_type # Creating the type column from their corresponding type
        # print(df.loc[type_mask, 'type'])
    df['type'] = df['type'].map(property_types) # Grouping the types into apartments, houses or plots  
    
    
extract_type(df_orig)
df_orig.head()
````

````python
def detect_outliers(df, senstivity, features):
    
    outliers = []
    cache = {} # Saving the data for labeling the outliers later
    
    # Iterating over the given columns and getting the first and third quartile of that column, then computing the interquartile range
    for feature in features:
        q1 = np.percentile(df[feature], 25)
        q3 = np.percentile(df[feature], 75)
        iqr = q3 - q1 
        
        # Senstivity can be tweaked for different results. Higher senstivity is less aggressive to capture outliers
        threshold = senstivity * iqr # This is the threshold which is going to be used in order to determine the value is an outlier or not
        
        lower_bound = q1 - threshold
        cache[feature + '_lower_bound'] = lower_bound
        upper_bound = q3 + threshold
        cache[feature + '_upper_bound'] = upper_bound
        # This will query all the rows that have a record less than q1 - threshold (too low) or q3 + threshold (too high) and return their indices
        outliers_col = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)].index
        
        outliers.extend(outliers_col) # Extending the found indices in this column to the outliers
        
    # If the record is classified as an outlier for both features, keeping it
    outliers = Counter(outliers) # Casting the list to Counter groups the items as keys and their occurences in the list as values, so we find values that occur twice in the outliers list
    multiple_outliers = list(k for k, v in outliers.items() if v > 1)
    
    return multiple_outliers, cache


def plot_outliers(normal_data, outlier_data, title, t=None):
    # This function is visualizing outliers over the data in order to label them overvalued or undervalued
    
    fig, ax = plt.subplots(figsize=(13, 13))
    
    area = 'area'
    if t == 'plot':
        area = 'plot_area'
    
    # Basically creating two subplots on top of each other, one of them has outliers and one of them has normal data
    plt.subplot(2, 1, 1)
    plt.scatter(x=normal_data[area], y=normal_data['price'], label='Normal Data')
    plt.title(title)
    plt.ylabel('Price', fontsize=13)
    plt.xlabel('Area', fontsize=13)
    
    plt.subplot(2, 1, 1)
    plt.scatter(x=outlier_data[area], y = outlier_data['price'], label='Outliers')
    plt.legend()
    
    plt.show()
    
    
def discrimination_plot(data, title, t=None):
    
    fig, ax = plt.subplots(figsize=(13, 7))
    
    area = 'area'
    if t == 'plot':
        area = 'plot_area'
    
    sns.scatterplot(x=data[area], y=data['price'], hue=data['outlier_type'], s=100)
    
    plt.legend(loc='lower right')
    plt.title(title)
    plt.ylabel('Price', fontsize=13)
    plt.xlabel('Area', fontsize=13)

    plt.show()   
    
````
